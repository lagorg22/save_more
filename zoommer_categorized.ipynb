```python
seen_ids = set()
seen_cats = set()
all_cats = []
for main_category in categories:
    for cat in main_category['childItems']:
        parent_cat_name = main_category['name']
        cat_name = cat['name']
        cat_id = cat['categoryId']
        if not cat_id or cat_id in seen_ids:
            continue
        if cat_name in seen_cats:
            cat_name = f'{parent_cat_name}_{cat_name}'
        seen_ids.add(cat_id)
        seen_cats.add(cat_name)
        all_cats.append([parent_cat_name, cat_name, cat_id])
```

```python
categories_cleaned = []
for parent_cat_name, cat_name, cat_id in all_cats:
    parent_cat_name = parent_cat_name.strip().lower().replace(' | ', '_').replace('\'', '').replace(' ', '_').replace('-', '_').replace('მონიტორები', 'monitors')
    cat_name = cat_name.strip().lower().replace(' | ', '_').replace('\'', '').replace(' ', '_').replace('-', '_').replace('მონიტორები', 'monitors')
    # if cat_name == 'type':
    #     cat_name = 'laptop_type'
    categories_cleaned.append((parent_cat_name, cat_name, cat_id))
categories_cleaned
```

```python
def get_response_data(cat_id):
    resp = requests.get(f'https://api.zoommer.ge/v1/Products/v3?CategoryId={cat_id}&Limit={random.randint(3422, 6543)}')
    resp = resp.json()
    df = pd.DataFrame(resp['products'])
    return df
```

```python
def clean_data(df):
    df_cl = df[['name', 'price', 'previousPrice', 'imageUrl', 'route']]
    df_cl.loc[:, 'route'] = df_cl['route'].apply(lambda x: f'https://zoommer.ge/{x}')
    df_cl.loc[:, 'price'] = df_cl['price'].astype(float) 
    df_cl = df_cl.rename(columns={'price': 'current_price', 'previousPrice': 'old_price', 'imageUrl': 'images', 'route': 'product_url'})
    df_cl['update_date'] = datetime.now()
    return df_cl
```

```python
engine = create_engine(database_url)
def export_data(df, db_name):
    df.to_sql(db_name, con=engine, if_exists='append', index=True)
```

```python
for main_cat, sub_cat, cat_id in categories_cleaned:
    try:
        db_name = f'{sub_cat}_zoom'
        df = get_response_data(cat_id)
        df = clean_data(df)
        export_data(df, db_name)
        print(f'Successfully exported data of {sub_cat}: {main_cat}: {cat_id}.')
    except Exception as e:
        print(f'\nError occured on {sub_cat}: {main_cat}: {cat_id} - {e}\n')
```
