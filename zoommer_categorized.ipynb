```python
resp = requests.get('https://api.zoommer.ge/v1/Content/mega-menu')
```

```python
categories = resp.json()
```

```python
categories[0]
```

```python
seen_ids = set()
seen_cats = set()
all_cats = []
for main_category in categories:
    for cat in main_category['childItems']:
        parent_cat_name = main_category['name']
        cat_name = cat['name']
        cat_id = cat['categoryId']
        if not cat_id or cat_id in seen_ids:
            continue
        if cat_name in seen_cats:
            cat_name = f'{parent_cat_name}_{cat_name}'
        seen_ids.add(cat_id)
        seen_cats.add(cat_name)
        all_cats.append([parent_cat_name, cat_name, cat_id])
```

```python
categories_cleaned = []
for parent_cat_name, cat_name, cat_id in all_cats:
    parent_cat_name = parent_cat_name.strip().lower().replace(' | ', '_').replace('\'', '').replace(' ', '_').replace('-', '_').replace('მონიტორები', 'monitors')
    cat_name = cat_name.strip().lower().replace(' | ', '_').replace('\'', '').replace(' ', '_').replace('-', '_').replace('მონიტორები', 'monitors')
    # if cat_name == 'type':
    #     cat_name = 'laptop_type'
    categories_cleaned.append((parent_cat_name, cat_name, cat_id))
categories_cleaned
```
