```python
import requests
import json
import pandas as pd
```

```python
resp = requests.get('https://api.zoommer.ge/v1/Content/mega-menu')
```

```python
categories = resp.json()
```

```python
all_cats = []
for main_category in categories:
    for cat in main_category['childItems']:
        all_cats.append([cat['name'], cat['categoryId']])
all_cats.sort(key=(lambda x: x[1] if x[1] else 0))
```

```python
cats_cleaned = []
seen_ids = set()
for cat in all_cats:
    if not cat[1] or cat[1] in seen_ids:
        continue
    seen_ids.add(cat[1])
    cats_cleaned.append(cat)
sorted(cats_cleaned)
cats_cleaned[1][0] = 'tv_monitor_accessories'
cats_cleaned[2][0] = 'laptop_accessories'
cats_cleaned[3][0] = 'photo_video_accessories'
cats_cleaned[8][0] = 'audio_systems'
cats_cleaned[9][0] = 'laptop_brands'
```
