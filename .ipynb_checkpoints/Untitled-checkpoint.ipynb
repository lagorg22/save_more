```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
# Specify Google Chrome binary path
chrome_options = Options()
chrome_options.binary_location = "/usr/bin/google-chrome"

# Initialize WebDriver
service = Service("/usr/bin/chromedriver")
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
search_term = input('Enter what are you searching for: ')
```

```python
driver.get(f'https://alta.ge/?match=all&subcats=Y&pcode_from_q=Y&pshort=Y&pfull=Y&pname=Y&pkeywords=Y&search_performed=Y&q={search_term}&dispatch=products.search&security_hash=1deb9f0f14f7d85d836c5f7f47c7461c')
```

```python
driver.get('https://alta.ge/?match=all&subcats=Y&pcode_from_q=Y&pshort=Y&pfull=Y&pname=Y&pkeywords=Y&search_performed=Y&q=iphone&dispatch=products.search&security_hash=1deb9f0f14f7d85d836c5f7f47c7461c')
```

```python
driver.close()
```
